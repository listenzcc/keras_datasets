```python
import keras
import random
import matplotlib.pyplot as plt
```

```python
datasets = dict(
    cifar10 = keras.datasets.cifar10,
    cifar100 = keras.datasets.cifar100,
    fashion_mnist = keras.datasets.fashion_mnist,
    mnist = keras.datasets.mnist,
#     reuters = keras.datasets.reuters,
)
for e in datasets:
    print(e)
    print(datasets[e].__doc__)
```

```python
plt.style.use('seaborn-dark')
for e in datasets:
    print('-' * 80)
    print(e)
    print(datasets[e].__doc__)
    (X_train, y_train), (X_test, y_test) = datasets[e].load_data()
    print(X_train.shape)
    for i in range(9):
        plt.subplot(3, 3, i+1)
        j = random.choice(range(X_train.shape[0]))
        plt.imshow(X_train[j], cmap='gray', interpolation='none')
        plt.title("Class {}".format(y_train[j]))
    plt.tight_layout()
    plt.savefig('%s.png' % e)
    plt.show()
```
